```python
def suan(s):#解码 获得图片的地址
    x = s[-1:] #  x = s.substring(s.length-1);
    w="abcdefghijklmnopqrstuvwxyz";
    xi=w.index(x)+1  # xi=w.indexOf(x)+1;
    sk = s[len(s)-xi-12:len(s)-xi-1] #sk = s.substring(s.length-xi-12,s.length-xi-1);
    s = s[:len(s)-xi-12] #s=s.substring(0,s.length-xi-12);
    k = sk[:len(sk)-1] # k=sk.substring(0,sk.length-1);
    f = sk[-1:]
    i = 0
    for kk in k:
        s=s.replace(kk,str(i))
        i+=1
    ss=s.split(f)
    s=""
    for sss in ss:
        s+=chr(int(sss)%256)
    return s
```
